```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
#Here is a case of Binary data
# drop the row with at least one element missing
#row with at least one element missing is for testing
df = pd.read_csv('heart.csv').dropna()
#change the column names to be a bit clearer and be easy to understand
df.columns = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 'cholesterol', 'fasting_blood_sugar', 'rest_ecg', 'max_heart_rate_achieved',
       'exercise_induced_angina', 'st_depression', 'st_slope', 'num_major_vessels', 'thalassemia', 'target']
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x = df.iloc[:,0:13]
y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 42)
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression(solver = 'lbfgs',max_iter=1000,random_state = 42)
logReg.fit(x, y)
```

```python
y_pred = logReg.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```
